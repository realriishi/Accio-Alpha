## Q1 – Rain on Two Days

At first I tried to list cases, but it got messy. The trick here is to use the complement.

If the probability that it **does not rain on either day** is 0.3, then the probability that it rains on at least one of the two days is just 1 − 0.3 = 0.7.

- Rain today **or** tomorrow: P = 0.7
- Rain today **and** tomorrow: 0.6 + 0.5 − 0.7 = 0.4
- Rain today but not tomorrow: 0.6 − 0.4 = 0.2
- Rain on exactly one day: 0.7 − 0.4 = 0.3

Once the union probability was clear, the rest followed pretty quickly.

## Q3 – Children Problem

Once one child is known to be a girl, the remaining children are still equally likely to be boys or girls.

So the probability that **all** children are girls depends only on the remaining (n−1) children:

(1/2)^(n−1).

This one felt simpler after thinking in terms of conditional probability.

## Q4 – Mixed Random Variable

Here the value of X comes from one of two distributions depending on a coin toss. I found it helpful to think of everything as a weighted average.

The CDF is:
F_X(x) = p·F_d(x) + (1 − p)·F_c(x).

The PDF and expectation follow the same idea — just combine the discrete and continuous parts using the coin probabilities.

## Q5 – Covariance

Z depends on X directly and also through the term X·Y². Since X and Y are independent standard normals, a lot of cross terms vanish.

After expanding carefully, the covariance works out to 2.

## Q6 – Job Offers

Instead of computing all the ways to get at least one offer, it’s much faster to compute the opposite case.

P(no offers) = (0.8)^4.
So P(at least one offer) = 1 − (0.8)^4 ≈ 0.59.

This made it clear that the 90% claim is way too optimistic.
