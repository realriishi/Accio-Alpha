## Q1
Given:
P(Rain today)=0.6, P(Rain tomorrow)=0.5, P(No rain on both days)=0.3.

P(Rain at least one day)=1−0.3=0.7.
P(Rain both days)=0.6+0.5−0.7=0.4.
P(Rain today only)=0.6−0.4=0.2.
P(Rain exactly one day)=0.7−0.4=0.3.
