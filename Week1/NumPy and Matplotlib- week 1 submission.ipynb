# Week 1 – NumPy and Matplotlib (Practice Notes)

These are my notes from experimenting with NumPy arrays and basic plots in Matplotlib. The goal was to get comfortable with indexing, shapes, and plotting syntax.

## NumPy Arrays

```

import numpy as np

# using a 2x3 matrix on purpose so I can practice row vs column indexing
numbers = np.array([[2, 4, 6],
                    [3, 7, 11]])

print(numbers)
print(numbers.shape)  # quick check: rows first, then columns
```

I picked a 2×3 matrix because it’s small enough to visualize, but not square — which helps avoid mixing up rows and columns.

## Random Values

```

# generating random values to see how NumPy fills the array
random_vals = np.random.rand(2, 3)
print(random_vals)
```

Re-running this cell gives different values each time, which helped me confirm that the randomness is working as expected.

## Line Plot

```

import matplotlib.pyplot as plt

time_points = [0, 1, 2, 3, 4]
temperature = [22, 21, 23, 22, 25]

plt.plot(time_points, temperature, color='darkorange')
plt.xlabel("Time")           # label x-axis so I don't mix it up later
plt.ylabel("Temperature")    # label y-axis clearly
plt.title("Temperature vs Time")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

I added a grid and title here mainly to make the plot easier to read. Earlier attempts failed when the x and y lists didn’t match in length.

## Scatter Plot

```

heights = [152, 159, 165, 173, 181]
weights = [54, 57, 66, 69, 82]

plt.scatter(heights, weights, color='skyblue')
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Height vs Weight")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

The points don’t fall on a perfect line, but the overall trend is still visible. This made it clearer why scatter plots are useful compared to line plots.
